### 1.4.
- Starting from 3801 and incrementing 3802, 3803... give each row an id number (in a new column)
- Based on monthly_income, 0-2500 as low, 2500-7500 as medium and above 7500 as high categorize it and add it as a new column with the name income_category.


```python
df["ID"] = range(3801, 3801 + len(df))
```

```python
df.head()
```

```python
def kategori(x):
  if x < 2500:
    return "low"
  elif x<=7500:
    return"medium"
  else:
    return "high"

df["INCOME_CATEGORY"] = df["AYLIK_GELIR"].apply(kategori)
df.head()
```
