```python
df = pd.read_csv("simple_data.csv")
```

```python
df.head()
```

```python
#df.groupby(['cat_col1']).agg.col.agg(['agg_func1', 'agg_func2'])
```

```python
meslek_grupby = df.groupby(['meslek']).aylik_gelir.agg(['mean']).head()
```

```python
#indeks kullanara yeni sütun yaratma
meslek_grupby.reset_index(inplace=True)
```

```python
meslek_grupby.head()
```

```python
#meslekleri ortalama kazançlarına göre sıralama
df.groupby(['meslek']).aylik_gelir.agg(['mean']).sort_values(by="mean", ascending = False).head()
```

```python
 #birden fazla niteliği gruplama ve birden g-fazla nitelik üzerinde kümeleme
 #fonksiyonu çalıştırma
 df.groupby(['meslek']).aylik_gelir.agg(['mean','sum','count']).sort_values("count", ascending=False).head()
```

```python
#birden fazla sıralama nasıl olur
df.sort_values(['yas','aylik_gelir'],ascending=False).head()
```

```python
#birden fazla nümerik deper üzerinde groupby()
df.groupby(["meslek"])[["yas","aylik_gelir"]].agg(["mean","sum","count"]).head()
```

```python
#çoklu sütun indeksnde sıralama (MultiIndex) #aylik gelir meana  gör sıraladık
df.groupby(["meslek"])[["yas","aylik_gelir"]].agg(["mean","sum","count"])\
.sort_values([('aylik_gelir','mean')], ascending=False) \
.head()
```
