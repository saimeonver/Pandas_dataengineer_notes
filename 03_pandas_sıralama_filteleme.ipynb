```python
df[df['meslek'] == 'Doktor'] #filtereleme
```

```python
df[(df['meslek'] == "Müzisyen")& (df['aylik_gelir']> 10000)]
```

```python
#query metodu ile sql where tarzı filtreleme
df.query("meslek == 'Doktor' | aylik_gelir > 10000").sort_values(by='aylik_gelir')
```
