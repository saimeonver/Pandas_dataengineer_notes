```python
df = pd.read_csv("OnlineRetail2.csv")
```

```python
df.head()
```

```python
#upper() country büyük harf

df.Country.str.upper().head()
```

```python
df['Country'] = df.Country.str.upper()
```

```python
df.head()
```

```python
#contains içerenler , startswith() ile başlayanlar

df[df.Description.str.contains("COFFEE",na=False)].head()
```

```python
df[df.InvoiceNo.str.contains("C",na=False)]
```

```python
df[df.InvoiceNo.str.startswith("C")].head()
```

```python
#replace yer değiştirme
#Bir sütun içinde geçen örüntüyü değiştirme
df['Description'] = df ['Description'].replace({'RETRO': 'COSTA'}, regex = True)
df[df.Description.str.contains("COFFEE",na=False)].head()
```

```python
#bir sütunda birden fazla örüntüyü değiştirme
df['Description'] = df ['Description'].replace({'RETRO': 'COSTA','MUGS':'MOCHA'}, regex = True)
df[df.Description.str.contains("COFFEE",na=False)].head()
```

```python
#nokta olarak bir hücreyi güncelleme (değer atama)
df.at[53,'StockCode']='37371'
df[df.Description.str.contains("COFFEE",na=False)].head()
```

```python
#sütunu truncate etmek
df["InvoiceDate"] = df["InvoiceDate"].apply(lambda x : x[:-9])
df[df.Description.str.contains("COFFEE",na=False)].head()
```
