SELECT

SELECT total_bill, tip, smoker, time
    FROM tips
    LIMIT 5;

```python
df[['total_bill','tip','smoker','time']].head(5)
```

    SELECT * FROM tips

    LIMIT 5;

```python
df[df['time'].str.contains('Dinner')].head(5)
```

YA DA

```python
df[df['time'] == 'Dinner'].head(5)
```

SELECT *
    FROM tips
    WHERE time = 'Dinner' AND tip > 5.00;

```python
df[ (df.time == 'Dinner' ) & (df.tip >5.00)].head(5)
```

 SELECT day, avg(total_bill) as total_bill, avg(tip) as tip
    FROM tips
    GROUP BY day;

```python
df.groupby(['day'])[['total_bill','tip']].mean().reset_index()
```

```python
df.groupby(['day'])[['total_bill','tip']].agg(['mean']).reset_index()
```

    SELECT day, AVG(tip), COUNT(*)
    FROM tips
    GROUP BY day;

```python
df.groupby('day')['tip'].agg(['mean','count']).head()
```

YA DA

```python
df.groupby(['day']).agg({'tip':np.mean, 'day' : np.size})
```

SELECT smoker, day, COUNT(*), AVG(tip)
    FROM tips
    GROUP BY smoker, day;

```python
df.groupby(['smoker','day'])['tip'].agg(['mean','count'])
```

YA DA


```python
df.groupby(['smoker','day']).agg({'tip':[np.size,np.mean]}).sort_values('day')
```
