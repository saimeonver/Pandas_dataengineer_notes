```python
#mevcut sütun degerini degiştirmek
df["aylikgelir"] = df["aylikgelir"] + 1000
```

```python
df.head()
```

```python
#yeni sütun eklemek
df["aylik_gelir_dolar"] = df["aylikgelir"] * 19.7
```

```python
df.head()
```

```python
#kategorik değişken üreterek yeni sütun oluşturmak

op_labels = ["low", "middle", "high"]
category = [0, 5000,8000, 100000]
df["gelir_seviyesi"] = pd.cut(df["aylikgelir"], bins=category, labels=op_labels,include_lowest=False)
```

```python
df.head()
```

```python
#sütunları birleştirmek
df["gelir_seviyesi"] = df["aylikgelir"].astype(str) + " - " + df["gelir_seviyesi"].astype(str)
```
