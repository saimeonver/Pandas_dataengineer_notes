```python
#dataframe okurken indeks sütununu belirlemek
#index_col indexbelirleyeceğim kolon, parse_dates ise datetime çevireceğim kolon

df = pd.read_csv("OnlineRetail2.csv", index_col="InvoiceDate",parse_dates=["InvoiceDate"])
```

```python
df.head()
```

```python
df.info()
```

```python
#dataframe i indekse göre sıralama
df = df.sort_index()
df.head()
```

```python
#tarih (Index) bilgisinden yeni nitelikller üretmek

df['year'] = df.index.year
df.head()
```

```python
#aynı bilgler index haricindeki stardart datetime sütunundan türetilebilir mi
df = df.reset_index() # Convert InvoiceDate from index to a regular column
df['InvoiceDate2'] = pd.to_datetime(df['InvoiceDate']) # This line will now work, but is redundant as reset_index() already preserves the datetime type
```
